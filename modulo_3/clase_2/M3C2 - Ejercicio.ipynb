```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Una compañía desea hacer predicciones del valor anual de sus ventas totales en cierto país a partir de la relación de éstas y la renta nacional. Para investigar la relación cuenta con los siguientes datos:
# X [189, 190, 208, 227, 239, 252, 257, 274, 293, 308, 316]
# Y [402, 404, 412, 425, 429, 436, 440, 447, 458, 469, 469]
# X representa la renta nacional en millones de euros e Y representa las ventas de la compañía en miles de euros en el periodo que va desde 2010 hasta 2020 (ambos inclusive). Calcular:
# a) La recta de regresión lineal de Y sobre X

var_x = np.array ([189, 190, 208, 227, 239, 252, 257, 274, 293, 308, 316])
var_y = np.array ([402, 404, 412, 425, 429, 436, 440, 447, 458, 469, 469])

var_linear_model = LinearRegression ()
var_linear_model.fit (var_x.reshape (-1, 1), var_y)

x_pred = np.linspace (var_x.min (), var_x.max ())
y_pred = var_linear_model.predict (x_pred.reshape (-1,1))

plt.scatter (var_x, var_y)
plt.plot (x_pred,y_pred)

plt.xlabel ('rentas')
plt.ylabel ('ventas')
plt.title ('Rentas vs Ventas')

plt.show ()
```

```python
# b) El coeficiente de correlación lineal e interpretarlo

var_correlation = var_linear_model.coef_ [0]

print (var_correlation)

var_interception = var_linear_model.intercept_

print (var_interception)

if var_correlation == 1:

    print ('correlacion positiva perfecta')

elif var_correlation > 0:

    print (f'correlacion positiva moderada = {var_correlation:.2f}')

elif var_correlation == 0:

    print ("no hay correlacion")

elif var_correlation < 0:

    print (f'correlacion negativa moderada = {var_correlation:.2f}')

elif var_correlation == -1:

    print ('correlación negativa perfecta')
```

```python
# c) Si en 2011 la renta nacional del país fue de 325 millones de euros. ¿Cuál será la predicción para las ventas de la compañía en este año?

var_prediction = var_linear_model.predict (np.array ([[325]]))

print (var_prediction)
```
