```python



import numpy as np
import matplotlib.pyplot as plt

# Definimos los parámetros de la distribución normal
mu = 4  # media
sigma = 2  # desviación típica

# Generamos una muestra de valores aleatorios de la distribución normal
num_samples = 1000
samples = np.random.normal(mu, sigma, num_samples)

# Graficamos un histograma de los valores aleatorios generados
plt.hist(samples, bins=30, density=True, alpha=0.5, color='blue')

# Generamos una serie de valores de x para graficar la función de densidad de probabilidad
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Graficamos la función de densidad de probabilidad
plt.plot(x, y, color='red')

# Ajustamos los ejes y agregamos etiquetas y título
plt.xlim(mu - 4*sigma, mu + 4*sigma)
plt.xlabel('Valores de x')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución normal con media 4 y desviación típica 2')

# Mostramos el gráfico
plt.show()
```

```python
import pandas as pd
from scipy.stats import norm

# Creamos un DataFrame con los valores de Z y las probabilidades acumuladas
z_values = pd.DataFrame({'Z': np.arange(0, 3.01, 0.2)})
z_values['Probabilidad'] = norm.cdf(z_values['Z'])

# a) P(0 ≤ Z ≤ 1.25)
prob_a = norm.cdf(1.25) - norm.cdf(0)
print("a) P(0 ≤ Z ≤ 1.25) =", prob_a)

# b) P(Z ≥ 1.25)
prob_b = 1 - norm.cdf(1.25)
print("b) P(Z ≥ 1.25) =", prob_b)

# c) P(Z ≤ -1.25)
prob_c = norm.cdf(-1.25)
print("c) P(Z ≤ -1.25) =", prob_c)

# d) P(0 ≤ Z ≤ 1.33)
prob_d = norm.cdf(1.33) - norm.cdf(0)
print("d) P(0 ≤ Z ≤ 1.33) =", prob_d)

# e) P(Z ≥ 1.33)
prob_e = 1 - norm.cdf(1.33)
print("e) P(Z ≥ 1.33) =", prob_e)

# f) P(-1.33 ≤ Z ≤ 0)
prob_f = norm.cdf(0) - norm.cdf(-1.33)
print("f) P(-1.33 ≤ Z ≤ 0) =", prob_f)

# Mostramos los valores de Z y las probabilidades acumuladas
print("\nValores de Z y probabilidades acumuladas:")
print(z_values)
```
