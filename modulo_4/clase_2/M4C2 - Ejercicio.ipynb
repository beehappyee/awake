```python
from scipy.stats import binom
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Caso 1: variable aleatoria en distribución normal
Si X es una variable aleatoria de una distribución normal con media μ y desviación estándar, hallar: P(μ - 3σ <= X <= μ  + 3σ)

```python
mean = 0
sigma = 1

x = np.linspace(mean - 3 * sigma, mean + 3 * sigma, 100)
y = np.exp(-(x - mean) ** 2 / (2 * sigma ** 2)) / (sigma * np.sqrt(2 * np.pi))

plt.plot(x, y)
plt.title('normal with mean 0 and sigma 1')
plt.xlabel('x')
plt.ylabel('probability')

x = 1.666
plt.axvline(x=-0.9974, color='green', linestyle='--')
plt.axvline(x=0.9974, color='green', linestyle='--')

plt.show()
```

Caso 2: distribución normal, media y desviación típica
En una distribución normal de media 4 y desviación típica 2, calcular el valor de a para que: P(4 - a <= x <= 4 + a) = 0.5934

```python
mean = 4
sigma = 2
samples = 1000
values = np.random.normal(mean, sigma, samples)

plt.hist(values, bins=30, density=True, alpha=0.5, color='blue')

x = np.linspace(mean - 4 * sigma, mean + 4 * sigma, samples)
y = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mean) ** 2 / (2 * sigma ** 2))

plt.plot(x, y, color='red')

plt.xlim(mean - 4 * sigma, mean + 4 * sigma)
plt.title('normal with mean 4 and sigma 2')
plt.xlabel('x')
plt.ylabel('probability')

a = 1.666
plt.axvline(x=4-a, color='green', linestyle='--')
plt.axvline(x=4+a, color='green', linestyle='--')

plt.show()
```

Caso 3: distribución normal
Dada una variable aleatoria continua Z, con distribución normal estándar, es decir, N(0;1), encuentre las siguientes probabilidades, usando la tabla
● P(0 ≤ Z ≤ 1,25)

```python
start = 0
end = 3
step = 1.33

z_values = pd.DataFrame({'Z': np.arange(start, end + step, step)})

z_values['probability'] = norm.cdf(z_values['Z'])

probability = norm.cdf(1.25) - norm.cdf(0)

print(f"P(0 ≤ Z ≤ 1,25) = {probability:.4f}")
```

● P(Z ≥ 1.25)

```python
probability = 1 - norm.cdf(1.25)

print(f"P(Z ≥ 1.25) = {probability:.4f}")
```

● P(Z ≤ -1.25)

```python
probability = norm.cdf(-1.25)

print(f"P(Z ≤ -1.25) = {probability:.4f}")
```

● P(0 ≤ Z ≤ 1.33)

```python
probability = norm.cdf(1.33) - norm.cdf(0)

print(f"P(0 ≤ Z ≤ 1.33) = {probability:.4f}")
```

● P(Z ≥ 1.33)

```python
probability = 1 - norm.cdf(1.33)

print(f"P(Z ≥ 1.33) = {probability:.4f}")
```

● P(-1.33 ≤ Z ≤ 0)

```python
probability = norm.cdf(0) - norm.cdf(-1.33)

print(f"P(-1.33 ≤ Z ≤ 0) = {probability:.4f}")
```

Caso 4: Distribución Binomial
Uno de los ítems de una prueba sorpresa consta de 12 preguntas de verdadero o falso. Suponiendo que los estudiantes no saben contestar a ninguna de ellas y, contestan al azar, hallar la probabilidad de obtener seis aciertos

```python
questions = 12
probability = .5

probability_6_t = binom.pmf(6, questions, probability)

print(f"probability_6_t = {probability_6_t:.4f}")
```

P: Distribución de probabilidad
R: Función que describe la probabilidad de ocurrencia de los posibles resultados de un evento aleatorio

P: Variables Aleatorias(discretas y Continuas)
R: Función que asigna un valor numérico a cada resultado posible de un experimento aleatorio. Si es discreta su valor toma valores aislados o contables, si es continua puede tomar cualquier valor dentro de un rango o intervalo específico

P: Distribución Normal
R: Distribuciones de probabilidad que se caracteriza por su forma de campana simétrica y su función de densidad de probabilidad en forma de curva

P: Distribución Binomial
R: Distribución de probabilidad discreta que modela el número de éxitos en una serie de ensayos independientes, donde cada ensayo tiene dos posibles resultados: éxito o fracaso
