5. Enuncie al menos tres diferencias entre el análisis factorial y el método de componentes principales

R: El análisis factorial y el método de componentes principales son dos técnicas de reducción de dimensionalidad que se utilizan para explorar y resumir la información contenida en un conjunto de variables. Aunque comparten algunos conceptos y objetivos similares, también presentan diferencias significativas. A continuación, se enuncian tres diferencias clave entre el análisis factorial y el método de componentes principales:

Objetivo y enfoque:

Análisis factorial: El objetivo principal del análisis factorial es identificar las relaciones subyacentes entre un conjunto de variables observadas y explicar la varianza en términos de factores latentes. Se basa en la idea de que las variables observadas son indicadores imperfectos de los factores subyacentes y busca encontrar una estructura de factores que explique la covarianza entre las variables.
Componentes principales: El objetivo principal del método de componentes principales es reducir la dimensionalidad del conjunto de datos al encontrar un nuevo conjunto de variables no correlacionadas llamadas componentes principales. Estas componentes principales son combinaciones lineales de las variables originales y se seleccionan de manera que capturen la mayor varianza posible en el conjunto de datos.
Naturaleza de las variables:

Análisis factorial: El análisis factorial se aplica generalmente a variables observadas que son medidas o variables manifiestas. Estas variables pueden ser categóricas o continuas, y se asume que están influenciadas por factores latentes no observables.
Componentes principales: El método de componentes principales se aplica a variables observadas que son medidas o variables manifiestas. Estas variables suelen ser continuas y se utilizan directamente para calcular las componentes principales.
Interpretación de los resultados:

Análisis factorial: El análisis factorial busca interpretar los factores latentes identificados en términos de una estructura subyacente. Los factores se interpretan en relación con las variables observadas y se buscan patrones o características comunes que expliquen la covarianza entre las variables.
Componentes principales: El método de componentes principales busca representar la varianza total del conjunto de datos mediante un conjunto reducido de componentes principales. Estos componentes no tienen una interpretación directa en términos de factores latentes, sino que son combinaciones lineales de las variables originales que maximizan la varianza capturada.
Es importante tener en cuenta que el análisis factorial y el método de componentes principales son técnicas complementarias y cada una tiene su aplicabilidad en diferentes situaciones. La elección entre ellos depende del objetivo del análisis y la naturaleza de los datos disponibles.

6. ¿Qué significa que el método de clusterización sea jerárquico?

R: Un método de clusterización jerárquico es aquel que crea una estructura de agrupamiento en forma de árbol o dendrograma, donde los elementos o grupos se van fusionando o dividiendo de acuerdo a su similitud. Este enfoque organiza los datos en una estructura jerárquica, donde los grupos pueden estar anidados dentro de otros grupos más grandes.

En un método de clusterización jerárquico, se pueden distinguir dos enfoques principales:

Agrupamiento aglomerativo (bottom-up): Inicia considerando cada elemento como un grupo individual y luego fusiona los grupos más cercanos gradualmente hasta formar un solo grupo que contiene todos los elementos. En cada paso, se decide qué grupos fusionar basándose en la similitud o distancia entre los grupos.

Agrupamiento divisivo (top-down): Comienza con un grupo que contiene todos los elementos y luego divide el grupo en subgrupos más pequeños de manera iterativa. En cada paso, se decide cómo dividir un grupo en función de la similitud o distancia entre los elementos dentro del grupo.

La principal ventaja de los métodos de clusterización jerárquicos es que proporcionan una representación visual del agrupamiento en forma de dendrograma, que muestra la relación de similitud entre los grupos y los elementos individuales. Esto permite una interpretación intuitiva de los resultados y la identificación de diferentes niveles de agrupamiento.

Sin embargo, los métodos de clusterización jerárquicos pueden ser computacionalmente más costosos que otros métodos, especialmente cuando se trabaja con conjuntos de datos grandes. Además, la división o fusión de grupos una vez realizada no se puede deshacer fácilmente, lo que limita la flexibilidad en la exploración de diferentes agrupamientos.

```python
dataframe = pd.read_csv('evaluacion_mkt_campaign.csv', sep=',')

describe = dataframe.describe()
print("distribution")
print(describe)
```

```python
# declarar ID como indice
dataframe.set_index('ID', inplace=True)

sales_cols = ['NumDealsPurchases','NumStorePurchases', 'NumWebPurchases', 'NumCatalogPurchases']
dataframe[sales_cols].describe()

# sumar las columnas de compras
dataframe['TotalPurchases'] = dataframe[sales_cols].sum(axis=1)
dataframe['TotalPurchases'].describe()
```

```python
# graficar boxplot de TotalPurchases y histograma
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].boxplot(dataframe['TotalPurchases'])
axs[0].set_title('Boxplot TotalPurchases')
axs[1].hist(dataframe['TotalPurchases'], bins=20)
axs[1].set_title('Histograma TotalPurchases')
plt.show()
```

```python
#filtrar el 5% de datos anomalos de TotalPurchases
q = dataframe['TotalPurchases'].quantile(0.95)
dataframe_filtered = dataframe[dataframe['TotalPurchases'] < q]
dataframe_filtered
```

```python
# graficar boxplot de TotalPurchases y histograma
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].boxplot(dataframe_filtered['TotalPurchases'])
axs[0].set_title('Boxplot TotalPurchases')
axs[1].hist(dataframe_filtered['TotalPurchases'], bins=20)
axs[1].set_title('Histograma TotalPurchases')
plt.show()
```

```python
# definir x e y con drop de TotalPurchases, Income
x = dataframe_filtered.drop(['TotalPurchases', 'Income'], axis=1)
y = dataframe_filtered['Income']
```

```python
# estandarizar los datos
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
pca = PCA(n_components=0.9)
pca.fit(x_scaled)
x_pca = pca.transform(x_scaled)

n_components = x_pca.shape[1]
```

```python
print(f'Number of components: {n_components}')
print(f'Explained variance: {pca.explained_variance_ratio_.sum()}')
```

```python
k_max = 8  # Número máximo de clusters a probar
wss = np.zeros(k_max)

for k in range(1, k_max + 1):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(x_pca)
    wss[k-1] = kmeans.inertia_
wss/=wss.max() # Normalizamos los valores de WSS

plt.figure()
plt.plot(range(1, k_max + 1), wss, color='black')
plt.scatter(range(1, k_max + 1), wss, color='blue')
plt.title("'Elbow Method' - data normalized")
plt.xlabel("Number of clusters K")
plt.ylabel("Total WSS")
plt.show()
```

```python
# range of cluster numbers to try
k_max = 8
min_clusters = 2
max_clusters = 8

# Sum of Squared Errors and silhouette scores
sse = np.zeros(k_max)
silhouette_scores = np.zeros(k_max)

for k in range(min_clusters, max_clusters + 1):

    print(f'Clustering with {k} clusters...')

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10,
    )

    kmeans.fit(x_pca)

    # Calculate SSE (Sum of Squared Errors)
    sse[k-1] = kmeans.inertia_

    # Calculate Silhouette score
    labels = kmeans.labels_
    score = silhouette_score(x_pca, labels)
    silhouette_scores[k-1] = score
```
