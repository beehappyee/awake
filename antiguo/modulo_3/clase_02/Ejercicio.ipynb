```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {
    'x': np.array([189,190,208,227,239,252,257,274,293,308,316]),
    'y': np.array([402,404,412,425,429,436,440,447,458,469,469])
}

model = LinearRegression()
model.fit(data['x'].reshape(-1,1), data['y'])

x_pred = np.linspace(data['x'].min(),data['x'].max())
y_pred = model.predict(x_pred.reshape(-1,1))

plt.scatter(data['x'], data['y'])
plt.xlabel('rentas')
plt.ylabel('ventas')
plt.plot(x_pred,y_pred)
plt.show()

```

```python
print('Intercepcion', model.intercept_)
print('Coef correlacion', model.coef_[0])
```
